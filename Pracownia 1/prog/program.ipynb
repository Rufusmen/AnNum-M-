```julia
#element_szeregu_Taylora(x,n) - oblicza (x^n)/n!
#sin_Taylor1(x) - oblicza sin(x) ze wzoru Taylora
#cos_Taylor1(x) - oblicza cos(x) ze wzoru Taylora
#sin_Taylor2(x) - oblicza sin(x) ze wzoru Taylora z przenawiasowaniem
#cos_Taylor2(x) - oblicza cos(x) ze wzoru Taylora
#sin_fraction(x) - oblicza sin(x) za pomocą ułamka łańcuchowego
#sin_mult(x) - oblicza sin(x) za pomocą iloczynu
#cos_mult(x) - oblicza cos(x) za pomocą iloczynu
```

```julia
using Plots
using Printf
include("program.jl")
```

```julia
n = collect(0.0:99.0);
n1= collect(-99.0:-1.0)
y=sin_Taylor1.(((pi*n)/Float64(100)));
y1=sin_Taylor1.(((pi*n1)/Float64(100)));
for i in 1:100
    y[i]=abs((sin(((pi*(i-1))/Float64(100))) - y[i])/sin(((pi*(i-1))/Float64(100))));
end
for i in 1:99
    y1[i]=abs((sin(((pi*(i-100))/Float64(100))) - y1[i])/sin(((pi*(i-100))/Float64(100))));
end
print(y, y1, "\n");
x=((pi*n)/Float64(100));
x1=((pi*n1)/Float64(100));

plot(x, y, color="red",  seriestype=:scatter, legend=false, xlabel="\$x\$", xlims = (0:pi), xticks = [-pi,-(pi/2),0,(pi/2),(pi)])
plot!(x1, y1, color="red",  seriestype=:scatter, legend=false, xlabel="\$x\$", xlims = (0:pi), xticks = [-pi,-(pi/2),0,(pi/2),(pi)])
```

```julia
savefig("sin_Taylor1.png");
```

```julia
n = collect(1.0:96.0);
n1= collect(-96.0:-1.0)
y=sin_Taylor1.(((pi*n)/Float64(100)));
y1=sin_Taylor1.(((pi*n1)/Float64(100)));
print(y1, "\n");
for i in 1:96
    y[i]=abs((sin(((pi*i)/Float64(100))) - y[i])/sin(((pi*i)/Float64(100))));
end
for i in 1:96
    y1[i]=abs((sin(((pi*(i-97))/Float64(100))) - y1[i])/sin(((pi*(i-97))/Float64(100))));
end
print(y1, "\n");
x=((pi*n)/Float64(100));
x1=((pi*n1)/Float64(100));

plot(x, y, color="red",  seriestype=:scatter, legend=false, xlabel="\$x\$", xlims = (0:pi), xticks = [-pi,-(pi/2),0,(pi/2),(pi)], ylims = (0,2*10^(-15)))
plot!(x1, y1, color="red",  seriestype=:scatter, legend=false, xlabel="\$x\$", xlims = (0:pi), xticks = [-pi,-(pi/2),0,(pi/2),(pi)], ylims = (0,2*10^(-15)))
```
